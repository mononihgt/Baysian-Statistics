```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_inline import backend_inline
backend_inline.set_matplotlib_formats('svg')
plt.rcParams['font.family'] = ['DejaVu Sans','simhei']
```

```python
class DistributionPlotter:
    def __init__(self):
        pass

    def Gamma(self, x, a, theta):
        return a * x**(a-1) * np.exp(-x**a / theta) / theta**a / gamma(a)

    def Beta(self, x, a, b):
        x = np.array(x, dtype=float)
        return x**(a-1) * (1-x)**(b-1) / beta(a, b)

    def plot_gamma(self, a, theta):
        x = np.linspace(0, 5, 1000)
        y = self.Gamma(x, a, theta)
        plt.plot(x, y)
        plt.title(f'Gamma Distribution a={a}, theta={theta}')
        # plt.ylim(0, 1)
        plt.show()

    def plot_beta(self, a, b):
        x = np.linspace(0, 1, 1000)
        y = self.Beta(x, a, b)
        plt.plot(x, y)
        plt.title(f'Beta Distribution a={a}, b={b}')
        plt.show()
```

```python
distributionPlotter = DistributionPlotter()
distributionPlotter.plot_gamma(5, 0.5)
```

```python
distributionPlotter = DistributionPlotter()
distributionPlotter.plot_beta(2,2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
x = np.linspace(0, 1, 1000)
y = distributionPlotter.Beta(x, 5, 7)
ax.plot(x, y, label='Beta(1,1)')
plt.legend()
ax.set_xlabel('p')
ax.set_ylabel('probiability density') 
# ax.set_ylim(0, 1.2)
plt.savefig('beta57.png',dpi=300)
plt.show()
```
